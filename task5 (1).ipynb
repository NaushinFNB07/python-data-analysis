```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('penguins cleaned.csv')
```

```python
sns.heatmap(df.corr(),annot=True)

```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation matrix')
plt.show()
```

```python
sns.pairplot(df[['Body Mass (g)', 'Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)']])
plt.show()
```

```python
sns.violinplot(x='Species',y='Body Mass(g)',data=df)
plt.figure(figsize=(10,5))
plt.title('Body mass by species')
plt.show()
```

```python
sns.swarmplot(x='Culmen Length (mm)',y='Culmen Depth (mm)',data=df)
plt.figure(figsize=(10,5))
plt.title('Culmen length and depth')
plt.show()
```

```python
sns.kdeplot(x='Body Mass (g)', data=df)
plt.show()
```
