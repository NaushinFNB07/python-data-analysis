```python
import numpy as np
arr1=np.array([1,2,3,4,5])
print("1D array:",arr1)
arr2=np.array([[1,2,3],[4,5,6]])
print("2D array:",arr2)
arr3=np.array([[[1,2],[3,4],[5,6]]])
print("3D array:",arr3)
a=np.array([10,20,30,40,50])
b=np.array([1,2,3,4,5])
print("addition:",a+b)
print("subtraction:",a-b)
print("multiply:",a*b)
print("divide:",a/b)
arr=np.array([1,2,3,4,5,6,7])
print("mean:",np.mean(arr))
print("sum:",np.sum(arr))
print("standard deviation:",np.std(arr))
new_arr=np.arange(1,10)
print("arange:",new_arr)
print("reshape:",new_arr.reshape(3,3))
print("linspace:",np.linspace(0,10,7))
print("indices of minimum values:",np.argmin(arr))
print("indices of maximum values:",np.argmax(arr))
```
