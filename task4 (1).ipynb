```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('penguins cleaned.csv')
```

```python
print(df.isnull().sum())
```

```python
sns.lineplot(x='Culmen Length (mm)',y='Culmen Depth (mm)',data=df)
plt.title('measurement of region')
plt.show()
```

```python
sns.scatterplot(x='Flipper Length (mm)',y='Body Mass (g)',data=df,color='green')
plt.title('Body mass calculation')
plt.show()
```

```python
sns.barplot(x='Body Mass (g)',y='Species',data=df,color='purple')
plt.title('Species body mass calculation')
plt.figure(figsize=(6,5))
plt.show()
```

```python
island_counts = df['Island'].value_counts()
plt.pie(island_counts, labels=island_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green'])
plt.title('Distribution of Penguins by Island')
plt.axis('equal')
plt.show()
```

```python
sns.histplot(data=df,x='Culmen Length (mm)',kde=True)
plt.title('distribution of culmen length')
plt.show()
```

```python
sns.boxplot(x='Species',y='Culmen Length (mm)',data=df,color='brown')
plt.title('species culmen length')
plt.show()
```
